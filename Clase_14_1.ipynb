```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Piloto": ['Max Verstappen', 'Lewis Hamilton', 'Charles Leclerc', 'Sergio Perez', 'George Russell'],
    "Puntos": [575, 234, 206, 285, 175],
    "Victorias": [19, 0, 0, 2, 0],
    "Podios" : [21, 6, 7, 9, 7],
    "Pole_Positions ": [12, 1, 5, 2, 0]
}
df = pd.DataFrame(data)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# --------------------------
# 1. Grfico de Barras
# --------------------------
ax1 = axes[0, 0]
colores_pilotos = ['blue', 'silver', 'red', 'darkblue', 'gray']
ax1.bar(df['Piloto'], df['Victorias'], color=colores_pilotos)
ax1.set_title('Grfico de Barras: Victorias por Piloto')
ax1.set_xlabel('Piloto')
ax1.set_ylabel('Nmero de Victorias')
ax1.grid(axis='y', alpha=0.5)
ax1.tick_params(axis='x', rotation=20)

# --------------------------
# 2. Grfico de Pastel
# --------------------------
ax2 = axes[0, 1]
ax2.pie(
    df['Puntos'],
    labels=df['Piloto'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colores_pilotos
)
ax2.set_title('Grfico de Pastel: Proporcin de Puntos Totales')
ax2.axis('equal')

# 3. Scatter Plot mejorado
ax3 = axes[1, 0]
ax3.scatter(df["Puntos"], df["Podios"], color='green', marker='o', s=180, alpha=0.8)
ax3.set_title("Scatter: Puntos vs. Podios")
ax3.set_xlabel('Puntos Totales')
ax3.set_ylabel('Nmero de Podios')

offsets = {
    "Max Verstappen": (10, -3),
    "Sergio Perez": (9, -4),
    "Lewis Hamilton": (9, -5),
    "Charles Leclerc": (10, -2),
    "George Russell": (5, 10)
}

for i, row in df.iterrows():
    piloto = row["Piloto"]
    dx, dy = offsets[piloto]
    ax3.annotate(
        piloto,
        (row["Puntos"], row["Podios"]),
        textcoords="offset points",
        xytext=(dx, dy),
        fontsize=9,
        ha='left'
    )

ax3.grid(axis='both', alpha=0.3)

# --------------------------
# 4. Histograma
# --------------------------
ax4 = axes[1, 1]
ax4.hist(df['Puntos'], bins=5, color='purple', edgecolor='black', alpha=0.7)
ax4.set_title("Grfico de Histograma: Distribucin de Puntos")
ax4.set_xlabel("Rangos de Puntos")
ax4.set_ylabel('Frecuencia (Nmero de Pilotos)')
ax4.grid(axis='y', alpha=0.5)

# Mostrar todo
plt.show()
```
